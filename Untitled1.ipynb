```python
#%% punto taller 9
import math
x1=(int)(input('(primer circulo)introduzca la coordenada en x: '))
y1=(int)(input('(primer circulo)introduzca la coordenada en y: '))
r1=(int)(input('(primer circulo)introduzca el radio: '))

x2=(int)(input('(segundo circulo)introduzca la coordenada en x: '))
y2=(int)(input('(segundo circulo)introduzca la coordenada en y: '))
r2=(int)(input('(segundo circulo)introduzca el radio: '))

#distancia desde la coord al centro para los dos circulos
a=(int)(input('coord en a: '))
b=(int)(input('coord en b: '))

dist1=math.sqrt((x1-a)**2+(y1-b)**2)
dist2=math.sqrt((x2-a)**2+(y2-b)**2)

cont=0
#solo sirve si distancia menor que el radio
#porlotanto esta dentro del circulo
if (dist1<=r1):
    print('(a,b) esta en el circulo 1')
    cont=cont+1
else:
    print('(a,b) no esta en el circulo 1')
if (dist2<=r2):
    print('(a,b) esta en el circulo 2')
    cont=cont+1
else:
    print('(a,b) no esta en el circulo 2')
if (cont==2):
    print('(a,b) esta en los dos circulos')
#cuando no entra a ninguno de los 2 if, entonces cont=0
if (cont==0):
    print('(a,b) no esta en ninguno de los dos circulos')
```

```python
#%% punto taller 18
tamano=(int)(input('ingrese el tamaño de la matriz cuadrada: '))
A=[] #se define la matriz
#se piden los valores los cuales se le asignaran a la matriz
for i in range(tamano):
    A.append([])
    for j in range (tamano):
        A[i].append((float)(input('numero para la posicion '+str(i+1)+','+str(i-1)+':')))
        
T=[]#se define la matriz triangular
cant1=[] #en este vector se guardan la cantidad de unos que hay en cada fila
#se llena la matriz superior hasta que las posiciones delos elementos en una 
#matriz superior esten ubicados cuando la fila es menor o igual que la columna
for i in range(tamano): 
    T.append([])
    for j in range(tamano):
        if i>j:
            T[i].append(0)
        else:
            T[i].append(A[i][j])
    #sirve para contar la cantidad de unos que hay en una fila
    cant1.append(T[i].count(1))
b=cant1.index(max(cant1))
a=cant1.count(max(cant1))
if max(cant1)==0:
    print('no hay unos en la matriz')
elif a==1:
    print('cantidad mayor de unos se encuentra en la fila ',b+1)
    
else: #si a!= entonces hay varias filas con el mismo maximo
    B=[] #aqui se guarda las posiciones del vector cant1 donde hay maximos
    for i in range(len(cant1)):
        if cant1[i]==max(cant1):
            B.append(i)
    print('las filas con mayor numero de unos son las siguientes: ',B)
        
```

```python
#%% punto taller 19
cant=(int)(input('Que cantidad de elementos desea en la lista: '))
Lista=[]
#Llenando la lista
for i in range(cant):
    Lista.append((input('Ingrese el elemento numero '+str(i+1)+': ')))
A=[]    #Matriz donde se guardara cada sublista de numeros consecutivos
tamano=[]   #Vector con los tamanos de cada sublista
for i in range(len(Lista)-1): 
    A.append([])    #Crea una sublista
    A[i].append(Lista[i]) #Le agrega el primer elemento analizado
    for j in range(i+1,len(Lista)):
        if Lista[j-1]<=Lista[j]:
            A[i].append(Lista[j])
        else:
            break
    tamano.append(len(A[i]))  #Introduce el tamaño de cada lista
cantMayor=tamano.count(max(tamano)) #Cuantas veces exise un maximo
pos=tamano.index(max(tamano)) #Posicion de la primera sublista mas larga
subLista=[]
if (cantMayor==1):
    print('La sublista mas larga ordenada ascendentemente es: ',A[pos])
else:
    a=-1    #a es un valor que dice la posicion desde donde se analizara el 
            #vector "tamano" para asi poder obtener los maximos
    for i in range(cantMayor):        
        a=tamano.index(max(tamano),a+1)            
        subLista.append(A[a])                
    print('Las sublistas mas largas ordenadas ascendentemente son: ',subLista)
```
